# Theta Sensitivity: Figure Demo

This notebook drives the plotting helpers in `plot_theta_sensitivity.py` and renders the artefacts inline. It loads `results/theta_sensitivity_4_16_100.pb`, aggregates trial outcomes by $(n, \theta)$, and reproduces:

1. an acceptance-rate heatmap,
2. median list size $|L|$ vs $\theta$ against the $4/\theta^2$ bound,
3. the precision/communication trade-off curve, and
4. the per-cell summary table.

```python
%matplotlib inline
import sys
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

import plot_theta_sensitivity as pts

# Make the plot module importable.
MODULE_DIR = Path.cwd() if Path.cwd().name == 'theta_sensitivity' else Path('results/figures/theta_sensitivity').resolve()
sys.path.insert(0, str(MODULE_DIR))
```

```python
def _show_inline(fig, name):
    print(f'[fig] {name}')
    plt.show()

pts.save = _show_inline
pts.setup_style()
```

## Load data and build per-cell aggregates

```python
data = pts.load_data()
trials = data['trials']
params = data['parameters']
thetas = sorted(params['thetaValues'])
ns = sorted(params['nRange'])

agg = pts.build_tables(trials)
print(f'{len(trials)} trials  |  n in {ns}  |  theta in {thetas}')
```

## Artefact: Acceptance heatmap

```python
pts.plot_acceptance_heatmap(agg, ns, thetas)
```

## Summary table

Read the CSV the module emits and display the first few rows.

```python
csv_path = MODULE_DIR / 'theta_sensitivity_summary.csv'
if not csv_path.exists():
    pts.write_summary_csv(agg, ns, thetas)

df = pd.read_csv(csv_path)
df.head(12)
```
